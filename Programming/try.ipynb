```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load audio file
audio_file = "E:\B.Tech  Project\Songs\Baarishein(PaglaSongs).mp3"
y, sr = librosa.load(audio_file)

# Display waveform
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform')
plt.show()

# Compute and display the spectrogram
D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)

plt.figure(figsize=(12, 8))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()

# Extract chroma feature
chroma = librosa.feature.chroma_stft(y=y, sr=sr)

# Display chromagram
plt.figure(figsize=(12, 4))
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plt.colorbar()
plt.title('Chromagram')
plt.show()

# Print features
print("Waveform shape:", y.shape)
print("Spectrogram shape:", D.shape)
print("Chroma feature shape:", chroma.shape)

```

```python


# Get tempo and beat frames
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)

# Print extracted features
print("Tempo:", tempo)
print("Beat frames:", beat_frames)
```

```python


# Get spectral centroid
centroid = librosa.feature.spectral_centroid(y=y, sr=sr)[0]

# Plot the spectral centroid over time
plt.figure(figsize=(12, 4))
plt.plot(librosa.times_like(centroid), centroid, label='Spectral Centroid')
plt.xlabel('Time (s)')
plt.ylabel('Spectral Centroid')
plt.title('Spectral Centroid over Time')
plt.legend()
plt.show()
```

```python


# Get MFCCs
mfccs = librosa.feature.mfcc(y=y, sr=sr)

# Display MFCCs
plt.figure(figsize=(12, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCCs')
plt.show()
```

```python


# Get zero-crossing rate
zero_crossings = librosa.feature.zero_crossing_rate(y)

# Plot zero-crossing rate over time
plt.figure(figsize=(12, 4))
plt.plot(librosa.times_like(zero_crossings), zero_crossings[0], label='Zero-Crossing Rate')
plt.xlabel('Time (s)')
plt.ylabel('Zero-Crossing Rate')
plt.title('Zero-Crossing Rate over Time')
plt.legend()
plt.show()
```

```python


# Get spectral contrast
spectral_contrast = librosa.feature.spectral_contrast(y=y, sr=sr)

# Display spectral contrast
plt.figure(figsize=(12, 4))
librosa.display.specshow(spectral_contrast, x_axis='time')
plt.colorbar()
plt.title('Spectral Contrast')
plt.show()
```

```python


# Get tonnetz
tonnetz = librosa.feature.tonnetz(y=y, sr=sr)

# Display tonnetz
plt.figure(figsize=(12, 4))
librosa.display.specshow(tonnetz, y_axis='tonnetz', x_axis='time')
plt.colorbar()
plt.title('Tonnetz')
plt.show()
```

```python


# Get spectral rolloff
rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)

# Plot spectral rolloff over time
plt.figure(figsize=(12, 4))
plt.plot(librosa.times_like(rolloff), rolloff[0], label='Spectral Rolloff')
plt.xlabel('Time (s)')
plt.ylabel('Spectral Rolloff')
plt.title('Spectral Rolloff over Time')
plt.legend()
plt.show()
```

```python


# Get harmonic component
harmonic = librosa.effects.harmonic(y)

# Plot harmonic component over time
plt.figure(figsize=(12, 4))
plt.plot(librosa.times_like(harmonic), harmonic, label='Harmonic Component')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Harmonic Component over Time')
plt.legend()
plt.show()
```

```python


# Calculate energy
energy = np.sum(librosa.feature.rms(y=y))

# Print extracted feature
print("Energy:", energy)
```

```python



# Calculate dynamic range
dynamic_range = np.max(y) - np.min(y)

# Print extracted feature
print("Dynamic Range:", dynamic_range)
```
